# Belief propagation for quantum LDPC codes, non binary belief propagation

In this notebook we finally get to stack classical methods so they could be used to correct quantum errors.

We will need the terminology we developed in the [linear algebra notebook](./linearAlgebra.ipynb), specifically, that there is an isomorphism between Pauli errors and binary vector-pairs:
$$P \rightarrow (P_z,P_x)$$

Then we observe that if we have matrices $H_x$ and $H_z$ as described in the [stabilizer codes and their constructions](./Stabilizer%20LDPC%20codes.ipynb), and define 
$$H = (H_z|H_x)$$
Which in numpy would be:
```python
H = np.hstack(H_z,H_x)
``` 
Then the syndrome of $P$ is defined as:
$$H_z \cdot P_x^T + H_x \cdot P_z^T$$
(note that the $X$ generator matrix detects the $Z$ component of $P$ and the $Z$ generator matrix detects the $X$ component of $P$).

This gives us a nice way to reduce the problem of decoding Pauli errors into two independent decoders, one defined by the $H_z$ matrix (which is fed $H_z \cdot P_x^T$), and one defined by the $H_x$ matrix (which is fed $H_x \cdot P_z^T$).

We get two (binary) error vectors $E_z$ and $E_x$, and combine them back to a Pauli error using:
$$(E_z,E_x) \rightarrow \tilde{P} $$

This works to some extent, but does not take into account any correlation between $x$ and $z$ errors, which we will cover later.

# Possible decoder outcomes
In the classical case, where we have a codeword $\vec{c}$ and an error $\vec{e}$ the decoder output could be one of the following:

1. A decoder failure. This is where the decoder is unable to recover $\vec{c}$ (or equivalently, $\vec{e}$), and is aware that it failed (by calculating $H\cdot \vec{c'} \neq \vec{0}$)

2. Decoder error. This is where the decoder beleives it found a solution, $\vec{c}$ (or equivalently, $\vec{e}$), where $H\cdot \vec{c'} = \vec{0}$, however, $\vec{c}\neq \vec{c'}$.

3. Decoder success. This is where the right solution was found.

Of the above three options, 3. is most desirable and 2. is most dangereous, and sometimes reffered to as false packet arrival (FPA) or silent data corruption (SDC).


Some literature jargon:
In [High-threshold and low-overhead fault-tolerant quantum memory](https://arxiv.org/abs/2308.07915)
A logical error is declared if $E$ and $E^*$ have the same syndrome, but they differ by a non identity logical operator.

From [Classical product code constructions for quantum Calderbank-Shor-Steane codes](https://quantum-journal.org/papers/q-2024-07-22-1420/pdf/)

We have two binary matrices, $H^Z$ which defines the $Z$ generators, and $H^X$, which defines the $X$ generators. 

Now, any Pauli operator $P$ defined by $(v^X,v^Z)$ that's made of just Xs (in some, not necessarily all indices) will commute with all the X generators of the code. 

It may, however, not commute with the Z generators of the code. This happens exactly when the number of overlapping indices between the where that operator has an X and (at least one of) a Z generator is odd, namely when:
$$H^X \cdot v^Z \neq 0 \ \ \ (modulu \ 2) $$

However, whenever we have  
$$H^X \cdot v^Z = 0 \ \ \ (modulu \ 2) $$
There are two options:

1. $v^Z$ either is a valid stabilizer, or a multiplication of a few of them (addition over $F(2)$) meaning, $v^Z \in span( (H^Z)^T )$
2. $v^Z$ is not in the stabilizer group of the code, namely: $v^Z \notin span( (H^Z)^T )$, in which it is a logical operator.

In the case that $v^Z \notin span( (H^Z)^T )$ and still $H^X \cdot v^Z = 0 \ \ \ (modulu \ 2) $, we say that $v^Z$ is a logical error.

We can check whether $v^Z$ is in $span((H^Z)^T)$ by appending it to $((H^Z)^T$ and using Gaussian elimination (assuming full rank to begin with).

From the paper:

""

For quantum CSS codes, one can distinguish different kinds of code minimum distance which we refer to as the pure distance $δ$ and the code distance $d$. The pure distances for X and Z errors are given, respectively, by 
$$\delta^x := min\{|v^x| : H^z \cdot v^x = 0, v^x \neq 0\}$$
$$\delta^z := min\{|v^z| : H^x \cdot v^z = 0, v^z \neq 0\}$$

and we simply call 
$$\delta := min(\delta^x, \delta^z)$$

the pure distance. By contrast, the code distances for X and Z errors are obtained by considering only errors patterns that do not leave the code-subspace invariant and thus do result in a logical error:
$$d^x := min \{|v^x| : H^z v^x = 0, v^x \in span((H^x)^T)\}$$
$$d^z := min \{|v^z| : H^x v^z = 0, v^z \in span((H^z)^T)\}$$

We call $d := min(d^x, d^z)$ the code distance. Note that we always have $d ≥ δ$ and $d$ can be significantly larger than $δ$ in certain cases.


""


Belief propagation generally works on factor graphs. In our context, we have a set of nodes that correspond to qubits, $V$, and another set of nodes that correspond to checks (factors in factor graph terminoplogy) $C$.

We will also have a set of edges $E$, that will connect nodes in $V$ to checks in $C$ (but not checks to checks or qubit nodes to qubit nodes). The set of edges, in our context, is given either by a parity check matrix, or a set of stabilizer generators.
When two checks $c,c'$ are both connected to $q,q'$, we will have a 4-cycle: $(q,c), (c,q'), (q',c'), (c',q)$.  


## Quaternary BP
The [algebra](/notebooks/linearAlgebra.ipynb) we covered includes handling of the case where we use the quaternary parity matrix $H$, and multiply it with the error vector $e$. Actually it's somewhat more elaborate than that, we start with a list of generators of the form $IIIZZ...$ ($Z$ generator) and $XXI...$ (an $X$ generator) and map them into $GF(4)$ vectors using:
$$I \rightarrow 0$$
$$X \rightarrow 1$$
$$Y \rightarrow \bar{\omega}$$
$$Z \rightarrow \omega$$

Then we can stack them as a matrix (which we can denote as $H$). By abuse of notation, we can talk about the elements of the matrix being $I,Z,X,Y$ instead of $0,1,\omega, \bar{\omega}$.

Then we have a Pauli error $E$, which we can map to a $GF(4)$ vector, $e$, in the same way. So remembering that multiplication in the Pauli group maps to $GF(4)$ addition, and that we want (!) the the equation:

$$s[i] = Tr(H\cdot e[i] = (H[i,0] \cdot e[0]) + (H[i,1] \cdot e[1]) ... + (H[i,n-1] \cdot e[n-1]))$$

To say $0$ if $E$ commutes with the $i'th$ stabilizer, and $1$ if it anticommutes, so we desinged it [in the algebra notebook](/notebooks/linearAlgebra.ipynb) to give us just that.





We now have most of the components to understand how BP over $GF(4)$ would look like.

Let's start with some minmal non trivial parity matrix:

$$H = 
\begin{bmatrix}
X & X & I\\
Z & Z & Z\\
\end{bmatrix}
$$

Or in $GF(4)$ notation:

$$H = 
\begin{bmatrix}
1 & 1 & 0\\
\omega & \omega & \omega\\

\end{bmatrix}
$$
Let's look at a check node, that corresponds to the first row in $H$
And an error
$$E = (X,I,I) \rightarrow (1,0,0)$$
That yields:
$$s[0] = Tr(H[0,:] \cdot E^T) = Tr(1 \cdot 1 + 1\cdot 0 + 0\cdot 0) = Tr(1) = 0$$
$$s[1] = Tr(H[1,:] \cdot E^T) = Tr(\omega \cdot 1 + \omega\cdot 0 + \omega\cdot 0) = Tr(\omega + 0 + 0) = Tr(\omega) = 1$$

And we are trying to find $E = (E_0,E_1,E2)$ with the same syndrome.

There are three variable nodes (corresponding to three columns of H, or three coordinates of $E$), and two check nodes.
The $E_i$ are variables, which we are trying to solve, and at a given moment, we estimate $E_i$ to be $I$ with probability $q_i^I$, $X$ with probability $q_i^X$, $Z$ with probability $q_i^Z$ and $Y$ with probability $q_i^Y$ (so their sum, $\sum_{_{W\in\{I,X,Y,Z\}}}q_i^W$ is $1$).

The variable nodes $E_i$ communicate these probabilities to each check node $C_j$ (not exactly, there are two subtlties here, but for now let's say to each check nodes).

At check node $C_0$ we are trying to assign confidence (probability, belief) in every one of the following equations:

Response to $E_0$ will be determined by:
$$Tr((1 \cdot E_1) + (1 \cdot E_2)) = s_0 = 0$$

Since $E_1$ and $E_2$ can be something from $\{I,X,Y,Z\}$, and $s_0 = 0$, the above equation represents the following $4\times4$ possible solutions, some of them are not valid because they do not equal $s_0=0$:
$$
\begin{aligned}
Tr((1\cdot 0) + (1 \cdot 0)) = 0 & \text{ with probability: } q_0^I \cdot  q_1^I \\  
Tr((1\cdot 0) + (1 \cdot 1)) = 0 & \text{ with probability:  } q_0^I \cdot  q_1^X \\
\cancel{Tr((1\cdot 0) + (1 \cdot \omega))} \neq 0& \text{ with probability:  } q_0^I \cdot  q_0^Z \\
\cancel{Tr((1\cdot 0) + (1 \cdot \bar{\omega}))} \neq 0 & \text{ with probability:  } q_0^I \cdot  q_1^Y \\
\end{aligned}
$$
In other words, if we fix (assume) $E_0$ to be $I$, then the probability of solving the constraint is $q_1^I + q_1^X$.

\

\
\

$$
\begin{aligned}
Tr((1\cdot 1) + (1 \cdot 0)) = 0& \text{ with probability:  } q_0^X \cdot  q_1^I \\
Tr((1\cdot 1) + (1 \cdot 1)) = 0& \text{ with probability: } q_0^X \cdot  q_1^I \\
\cancel{Tr((1\cdot 1) + (1 \cdot \omega))} \neq 0 & \text{ with probability: } q_0^X \cdot  q_1^Z \\
\cancel{Tr((1\cdot 1) + (1 \cdot \bar{\omega}))} \neq 0 & \text{ with probability:  } q_0^X \cdot  q_1^Y \\
\end{aligned}
$$
In other words, if we fix (assume) $E_0$ to be $X$, then the probability of solving the constraint is $q_1^I + q_1^X$.
\

This allows us to understand the ominous Equation (10) in [Refined Belief Propagation Decoding of Sparse-Graph Quantum Codes](https://arxiv.org/abs/2002.06502) (which, actually is a paper that explains it well):

"Horizontal Step: At check node m, compute $r_{mn} = (r^I_{mn}; r^X_{mn}; r^Y_{mn}; r^Z_{mn})$ and pass rmn as the message $m \rightarrow n$ for every $n \in N(m)$, where"
$$r^W_{mn} = \sum\limits_{E|_N(m): E_n=W, \langle E|_{N(m)}, S_m|_{N(m)}\rangle = z_m} (\prod\limits_{n'\in N(m)\setminus n} q_{mn'}^{E_{n'}})$$

For $W\in \{I,X,Y,Z\}$

The crucial part to understand is the condition for summation under the $\sum$ sign, i.e.:
$$E_n=W, \langle E|_{N(m)}, S_m|_{N(m)}\rangle = z_m$$

Except in our notation so far we used $H[m,:]$ instead of $S_m$ and $s_m$ instead of $z_m$, as well as $Tr(x,y)$ instead of $\langle , \rangle$. So this equation takes the form:

$$E_n = W, Tr(H[m,:],E) = s_m$$

So fix the $n$'th componenet of the error pattern to be $W$ (where $W \in \{I,X,Y,Z\}$), and check which assignments of the rest of the components of $E$ could solve the equation. For each assignment that solves the equation, multiply the probabilities for the individual components - this gives one component in the summation.

Although the equations only need to be solved once at the beginning of life of the code (!) (not even the decoder), this implies testing $4^d$ where $d$ is the number of non zero elements in $H$ which limits the scalability of this "list" type decoder.


````python
from qecc.gf4 import *
from qecc.polynomialCodes import codes
from itertools import permutations, product
from line_profiler import profile


def generateSolutionsArray(array):
    """
    array is assumed to be a one dimensional array over the set {0,1,2,3}
    """
    gf4Ints = [0,1,2,3]
    GF4 = [gf4(i) for i in gf4Ints]
    solutions = {0: [], 1: []}
    #if not isinstance(array,np.ndarray):
    #    raise ValueError("array must be an ndarray")
    nonZeros = np.where(array != 0)[0]
    arrayAsGf4 = list(map(gf4, array[nonZeros]))
    print(f"There are {len(nonZeros)} non zero entries in the array, at indices {nonZeros}")
    candidates = product(GF4, repeat=len(nonZeros))
    for c in candidates:
        result = traceInner(arrayAsGf4,c)
        solutions[result.value].append(toarray(c))
    return solutions


def generateSolutions(H):
    """
    H is assumed to be a two dimensional array over the set {0,1,2,3}
    """
    solutions = {}
    #if not isinstance(H,np.ndarray):
    #    raise ValueError("H must be an ndarray")
    for row in range(H.shape[0]):
        print(f"Generating solutions for row {row} of H")
        solutions[row] = generateSolutionsArray(H[row,:])
    return solutions



````

# Methodology
So here is the workflow for evauluating a quantum LDPC code using the depolarizing channel:
Arguments:
Hx, Hz: A pair of binary matrices. Use the codes in polynomialCodes.py
errorRange: A range of probabilities

Returns:
logicalErrorRate: array of floats. For each error probability p, the number of times a logical error was created divided by the number of samples attempted for the error probability.
decoderFailureRate: array of floats. For each error probability p, the number of times the decoder failed divided by the number of samples attempted for the error probability.

```python
import numpy as np
from qecc.gf4 import integerTraceProduct as tp
from qecc.gf4 import integerToDualBinary
from qecc.logicals import computeLogicals
seed = 7134066
localRandom = np.random.RandomState(seed)
logicalX, logicalZ = computeLogicals(Hx, Hz)
```


1. Loop over error probabilities p in some range, say:
```python
for p in np.linspace(0.0003, 0.1, 50):

```
2. Sample (some number of times) an error, which is a vector over {0,1,2,3} representing I,X,Z,Y (to be consistent check with the documentation in gf4.py)
```python
    error = localRandom.choice([0,1,2,3], size=H.shape[1], replace=True, p=[1 - 3*p, p, p, p])
```
3. Calculate the syndrome for this error
```python
    syndrome = np.array([(np.sum(list([map(tp, H[i,:], E)]))%2) for i in range(H.shape[0])])
```
4. Run whatever decoder you decided to evaluate with arguments Hx, Hz, syndrome. The decoder has to accept a binary pair and syndrome, and return errorEstimation (an estimated solution), which is an array over 0,1,2,3 and a success flag (boolean).
```python
    estimatedError, success = decoder(Hx, Hz, syndrome)
```
5. If the decoder thinks it failed, add 1 to the decoderFailure counter. 

```python
    decoderFailureRate = decoderFailureRate + (not success)
```
6. If the decoder thinks it succeeded (success == True), then test the residual error, i.e.: the (gf(4)) sum of the estimated error and the original error, to see if either: 

    6.1. It does not amount to a 0 syndrome (this means the decoder is wrong, meaning - it thinks that the estimatedError + error is a stabilizer, and that it countered the effect of the error up to a stabilizer, but it didn't, and the result is not in the normalizer).

    6.2. It is a logical error.

In either case we increase the decoderErrorRate by 1.

Note that, although decoder failure is usually counted with decoder error in most papers, we are separating it here.

Checking whether the residual error gives $0$ syndrome means it commutes with all stabilizers. By definition, this places it in the normalizer of $S$ i.e.: $$N(S)$$
A valid solution $\hat{E}$, is such that the residual error is a stabilizer. I.e., when we add (over $GF(4)$) the residual error $\hat{E}$ to the actual error $E$, we get that
$$\hat{E} \oplus E \in \langle S \rangle$$

We already checked that $\hat{E} \oplus E \in N(S)$. If it *is* a stabilizer, then it commutes with the logical operators that we computed in advance. So that's all we need to check. We have the table below that summarises this for the residual $Z$ error:

|               | $Hx \cdot residualErrorZ == 1 (mod2)$ | $Hx \cdot residualErrorZ == 0 (mod2)$ |
| :------------ | :----------- | :------------ |
| $logicalX \cdot residualErrorZ == 1 (mod2)$ | ❌Not in the normalizer (and in particular not in the stabilizer) | ❌ Not in the stabilizer |
| $logicalX \cdot residualErrorZ == 0 (mod2)$ | ❌Not in the normalizer (and in particular not in the stabilizer)| ✅ Is in the normalizer *and* is in the stabilizer. |

```python
    errorX, errorZ = integerToDualBinary(error)
    estimatedErrorX, estimatedErrorZ = integerToDualBinary(estimatedError)
    
    residualErrorX = (estimatedErrorX + errorX) %2
    residualErrorZ = (estimatedErrorZ + errorZ) %2
    # Check whether the residual error gives 0 syndrome:
    if not ( np.all((np.dot(Hx,residualErrorZ)) % 2 ==0) and np.all((np.dot(Hz, residualErrorX)%2) == 0)):
            decoderError[p] += 1
    else: # So we are in the case that the residual error commutes with all stabilizers, i.e., it is in the normalizer. So let's check if it is a stabilizer (commutes with all logicals), or a logical error (anticommutes with some logical operator)
        if not ( np.all((np.dot(logicalZ,residualErrorX) % 2 )== 0) and np.all((np.dot(logicalX, residualErrorZ) % 2)==0)):
            decoderError[p] += 1
```
