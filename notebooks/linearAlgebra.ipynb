# Linear algebra
We will be dealing with linear algebra over $\mathbb{C}$. 

Excersize:
Show that if $V,W$ are vector spaces with bases $A,B$ respectively, then $A \otimes B = \{a \otimes b : a \in A, b \in B\}$ is a basis for $V \otimes W$. Conclude that if $V, W$ are finite dimensional, then the dimension of $V \otimes W$ is $\dim(V) \cdot \dim(W)$.

Excersize:
Show that the matrices $I = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}$, $X = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix}$, $Z = \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix}$ and $Y = \begin{pmatrix} 0 & -1 \\ 1 & 0 \end{pmatrix}$ are:
1. Linearly independent.
2. Span the space of 4 dimensional matrices over $\mathbb{C}$ (hint: what is the dimension of 2X2 matrices as a vector space ?)
3. Form a group under matrix multiplication.
4. Confirm that $X^2 = Y^2 = Z^2 = I$
5. Confirm that $X\cdot Z = Y$ and that $Z \cdot Y = X$
6. Orthogonal to each other (as 4 dimensional vectors over $\mathbb{C}$)

Excersize:
For $P_1, P_2$, which are two dimension $n$ Pauli operators, i.e. a member of ${X_1 \otimes X_2 \otimes ... X_n : X_i \in \mathbb{P}}$ prove that $P_1, P_2$ commute if the number of indecies $i$ where $X_i$ and $Y_i$ anticommute is even.

Let $P$ be a dimension $n$ Pauli operator $P = X_1 \otimes X_2 \otimes ... X_n : X_i \in \mathbb{P}$ and define two vectors: $P_x$ and $P_z$ by: 

$P_x(i) = 0 = P_z(i)$ if $P(i)=I$,

$P_x(i) = 1$ if $P(i)=X$,

$P_z(i) = 1$ if $P(i)=Z$ and

$P_x(i) = 1 = P_z(i)$ if $P(i)=Y$

Excersize:
Show that if $P,Q$ are two $n$ dimension Pauli operators, then the decomposition above respects multiplication. i.e., if $P\cdot Q=S$ then:

$S_x = P_x + Q_x mod 2$

$S_y = P_y + Q_y mod 2$
